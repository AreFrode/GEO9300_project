### *Scientific report structure:
Your report should contain at least the following parts. You can write additional sections as well.

**1. Introduction:**

Write what is the aim and content of the report. Typically you have one or two research questions you want to answer.

**2. Study area:** 

Visualize and decribe the study area relevant for your analysis. Examples of relevant content are a map including dots showing the locations of the data collection, and general climatological/hydrological/geophysical characteristics of the region or more directly related to the data.

**3. Data:** 

In this section, you should describe and visualize the data, quality, resolution, period etc, and do some basic data analysis. For example you can calculate and visualize simple statistics such as means, standard deviations, annual time series, how the data varies over the year or among themselves, etc.

**4. Methods:** 

Here you should present the methods you are using in the statistical analysis. Show formulas and explain every abbreviation you use and include unit of all variables. If you want to do som modifications and changes in the data you use for your analyses, explain what you have done and why.
       
**5. Results and discussion:**

In this section, you should do the statistical analyses reflecting the aim of your project. Visualize your findings in a smart and easy-to-understand manner. Describe and discuss your findings.

**6. Summary and conclusion:** 

Write a short summary and conclusion based on your results, reflecting the aim.

**7. Resources:**

Write a short comment on where to find the code (link to a github repository). If you cannot publish your code openly availible, write a statement about why and upload the code as a zip-file to the assignment delivery.

### Report Tips:
 - **Maximum 12 publication units (#words/500 + #figures + #tables). No appendix or supplement. So be concise!**
