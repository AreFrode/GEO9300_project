```python
import pandas as pd
import numpy as np
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
path_data = "../dataset/"
buoys_df = pd.read_csv(f"{path_data}prepared_buoy_data.csv", index_col=[0, 1])

kvs = buoys_df.loc[['KVS_SvalMIZ_03',
                    'KVS_SvalMIZ_07']].dropna().reset_index(level=0)

kvs_test = buoys_df.loc['KVS_SvalMIZ_10'].dropna()
```

```python
time_numeric = pd.to_datetime(kvs.index).astype(np.int64) / 10 ** 9
```

```python
fig, ax = plt.subplots(ncols = 2, figsize = (8, 6), sharey = True)

sns.scatterplot(data = kvs, x = 'arome_t2m', y = 'temp_air', hue = kvs.index, ax = ax[0], legend = False)
ax[0].set(xlabel = 'Forecasted T2M', ylabel = 'Observed T1M')
scatter = sns.scatterplot(data = kvs, x = 'sic', y = 'temp_air', hue = kvs.index, ax = ax[1], legend = False)
ax[1].set(xlabel = 'Reanalysis SIC')

norm = plt.Normalize(time_numeric.min(), time_numeric.max())
```

```python
fig, ax = plt.subplots(ncols = 2, figsize = (8, 6), sharey = True)

sns.scatterplot(data = kvs_test, x = 'arome_t2m', y = 'temp_air', hue = kvs_test.index, ax = ax[0], legend = False)
ax[0].set(xlabel = 'Forecasted T2M', ylabel = 'Observed T1M')
scatter = sns.scatterplot(data = kvs_test, x = 'sic', y = 'temp_air', hue = kvs_test.index, ax = ax[1], legend = False)
ax[1].set(xlabel = 'Reanalysis SIC')
```
